```python
import tensorflow as tf # Import tensorflow library
import matplotlib.pyplot as plt # Import matplotlib library
import numpy as np
```

## Download dataset

```python
mnist = tf.keras.datasets.mnist # Object of the MNIST dataset
(x_train, y_train),(x_test, y_test) = mnist.load_data() # Load data
```

```python
plt.imshow(x_train[0], cmap="gray") # Import the image
plt.show() # Plot the image
```

```python
# Normalize the train dataset
x_train = tf.keras.utils.normalize(x_train, axis=1)
# Normalize the test dataset
x_test = tf.keras.utils.normalize(x_test, axis=1)
```

### Activation function of Layers =  ùë°ùëéùëõùêª , sùëúùëìùë°ùëöùëéùë•

```python
#Build the model object
model1 = tf.keras.models.Sequential()
# Add the Flatten Layer
model1.add(tf.keras.layers.Flatten())
# Build the input and the hidden layers
model1.add(tf.keras.layers.Dense(128, activation=tf.nn.tanh))
model1.add(tf.keras.layers.Dense(128, activation=tf.nn.tanh))
# Build the output layer
model1.add(tf.keras.layers.Dense(10, activation=tf.nn.softmax))
```

```python
# Compile the model
model1.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
```

```python
model1.fit(x=x_train, y=y_train, epochs=5) # Start training process
```

```python
# Evaluate the model performance
test_loss, test_acc = model1.evaluate(x=x_test, y=y_test)
```

```python
# Print out the model accuracy 
print('\nTest accuracy:', test_acc)
```
